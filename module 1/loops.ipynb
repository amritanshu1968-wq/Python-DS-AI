```python
#print all numbers from 1 to 50 that are divisible by 3
for i in range(1, 51):
    if i % 3 == 0:
        print(i)
```

```python
#use a while loop to compute the factorial of a number entered by the user

num = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= num:
    factorial *= i
    i += 1
print(f"The factorial of {num} is {factorial}")
```

```python
#wap that prints the fibonacci sequence up to n terms using a loop
num_terms = int(input("Enter the number of terms: "))
n1, n2 = 0, 1
print("Fibonacci sequence:")
for i in range(num_terms):
    print(n1)
    nth = n1 + n2
    n1 = n2
    n2 = nth
```

```python
#Use a list comprehension to create a list of all vowels found in a given sentence.
sentence = input("Enter a sentence: ")
vowels = [char for char in sentence if char.lower() in 'aeiou']
print("Vowels in the sentence:", vowels)
```

```python
# Write a number-guessing game: the program picks a random number between 1 and 100 (use the random module), and the user keeps guessing until correct. After each guess, print "Too high" or "Too low". Track and print the number of attempts at the end.
import random

# Generate a random number between 1 and 100
number = random.randint(1, 100)

attempts = 0

print("Guess the number between 1 and 100")

while True:
    guess = int(input("Enter your guess: "))
    attempts += 1

    if guess > number:
        print("Too high")
    elif guess < number:
        print("Too low")
    else:
        print("Congratulations! You guessed the correct number.")
        break

print("Total attempts:", attempts)
```
