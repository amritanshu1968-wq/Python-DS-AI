```python
#create a variable for your favourite movie title(str), release year(int), rating out of 10
#recommend it bool(true/false) print each type

# Create variables
movie_title = "3 Idiots"      # String
release_year = 2009           # Integer
rating = 9.5                  # Float
recommend = True              # Boolean

# Print values
print("Movie Title:", movie_title)
print("Release Year:", release_year)
print("Rating:", rating)
print("Recommend:", recommend)

# Print data types
print("\nData Types:")
print(type(movie_title))
print(type(release_year))
print(type(rating))
print(type(recommend))
```

```python
#Create a dictionary representing a product with keys
#name,price,and tags (a list). write code to dd a new tag,
#update the price by applying a 10% discount, and print a formatted sentence
#describing the final product using an f-string.

# Create a dictionary representing a product
product = {
    "name": "Laptop",
    "price": 50000,
    "tags": ["Electronics", "Portable"]
}

# Add a new tag
product["tags"].append("Discounted")

# Apply a 10% discount
product["price"] = product["price"] * 0.9  # 10% discount

# Print the final product details using an f-string
print(f"The product '{product['name']}' now costs ₹{product['price']:.2f} and belongs to the categories: {', '.join(product['tags'])}.")

```

```python
#Bitwise Operators

#bitwise operators demo
a = 5  # binary: 0101
b = 3  # binary: 0011
print("AND =", a & b)
print("OR =", a | b)
print("XOR =", a ^ b)
print("NOT =", ~a)
print("Left Shift =", a << 1)
print("Right Shift =", a >> 1)

#mixed Expression
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
result = (a + b) * c - a / b
print("Result =", result)
```

```python
#Assignment Operators
#Assignment operators Demo

a = 10

a += 5
print("After += :", a)

a -= 3
print("After -= :", a)

a *= 2
print("After *= :", a)

a /= 4
print("After /= :", a)

a %= 3
print("After %= :", a)
```

```python
#given two numbers entered by user, write expression to print their sum, difference
#product quotient, floor division, remainder, and the first number raised to the power of the second number.
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Sum =", num1 + num2)
print("Difference =", num1 - num2)
print("Product =", num1 * num2)
print("Quotient =", num1 / num2)
print("Floor Division =", num1 // num2)
print("Remainder =", num1 % num2)
print("Power =", num1 ** num2)
```
