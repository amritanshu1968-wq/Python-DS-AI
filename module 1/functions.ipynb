```python
def add(a,b):
    return a + b
result = add(5, 3)
print(result)
```

```python
def subtract(a,b):
    return a - b
result = subtract(10, 4)
print(result)
```

```python
def multiply(a, b):
    return a * b
result = multiply(5, 3)
print(result)
```

```python
def divide(a, b):
    return a / b
result = divide(15, 3)
print(result)
```

```python
def quotient(a, b):
    return a % b
result = quotient(20, 8)
print(result)
```

```python
def square(n):
    return n ** 2
print(square(25))
print(square(12))
```

```python
def add_all(*args):
    return sum(args)
print(add_all(23,25,14,15,16))
```

```python

def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Amritanshu", age=21, city="Lucknow")
```
