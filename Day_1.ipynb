```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Name": ["Amritanshu", "Shivam", "Rohit", "Amit", "Ayush", "Gaurav"],
    "Age": [21, 22, 23, 24, 25, 26],
    "City": ["Lucknow", "Kanpur", "Delhi", "Mumbai", "Jaunpur", "Varanasi"],
    "Hours_Study": [5, 6, 7, 8, 2, 9],
    "marks": [90, 80, 70, 60, 50, 40]
}
```

```python
#Covert data into a DataFrame
df = pd.DataFrame(data)
```

```python
#Display full dataset
print("Full Dataset")
print (df)
```

```python
#head (first 5 rows)
print("\nFirst 5 rows using head()")
print (df.head())
```

```python
#Tail (last 5 rows)
print("\nLast 5 rows using tail()")
print (df.tail())
```

```python
#dataset information
print("\nDataset Information")
print (df.info())
```

```python
#statistical Description
print("\nStatistical Description")
print (df.describe())
```

```python
#mean of marks
print("\nMean of marks")
print (df['marks'].mean())
```

```python
#maximum marks
print("\nMaximum marks")
print (df['marks'].max())
```

```python
#minimum marks
print("\nMinimum marks")
print (df['marks'].min())
```

```python
#filter students who scored more than 60 marks
print("\nStudents who scored more than 60 marks")
print (df[df['marks'] > 60])
```

```python
#correlation between marks and hours of study
print("\nCorrelation")
print (df[["Hours_Study", "marks"]].corr())
```

```python
#plot gragh
plt.scatter(df['Hours_Study'], df['marks'])
plt.title('Hours of Study vs Marks')
plt.xlabel('Hours of Study')
plt.ylabel('Marks')
plt.show()
```
